<h2 style="color:DodgerBlue">Описание проекта:</h2>

----
Описание задачи:
Создать базовый класс Employee в C#, который будет представлять информацию о
сотрудниках компании. На основе этого класса разработать 2-3 производных
класса, демонстрирующих принципы наследования и полиморфизма. В каждом из
классов должны быть реализованы новые атрибуты и методы, а также
переопределены некоторые методы базового класса для демонстрации
полиморфизма.
Требования к базовому классу Employee:
• Атрибуты: Имя (Name), Возраст (Age), Зарплата (Salary).
• Методы:
o GetInfo(): метод для получения информации о сотруднике в виде
строки.
o Work(): метод для вывода сообщения о выполнении рабочих
обязанностей.
o TakeVacation(): метод для вывода сообщения о взятии отпуска.

Требования к производным классам:
1. Менеджер (Manager): Должен содержать дополнительные атрибуты, такие
как Количество подчиненных (SubordinatesCount). Метод Work() должен быть
переопределен для добавления информации о управлении командой.
2. Разработчик (Developer): Должен содержать дополнительные атрибуты,
такие как Специализация (Specialization). Метод TakeVacation() должен быть
переопределен для добавления информации о необходимости согласования
отпуска с проектными сроками.
3. Дизайнер (Designer) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Стиль дизайна (DesignStyle).
Метод Work() должен быть переопределен для добавления информации о
работе над дизайн-проектами.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

<h2 style="color:DodgerBlue">Реализация:</h2>

----

```
using System;

// Базовый класс Employee
class Employee
{
    private string name;
    private int age;
    private decimal salary;
    private string department;
    private string position;

    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    public int Age
    {
        get { return age; }
        set { age = value; }
    }

    public decimal Salary
    {
        get { return salary; }
        set { salary = value; }
    }

    public string Department
    {
        get { return department; }
        set { department = value; }
    }

    public string Position
    {
        get { return position; }
        set { position = value; }
    }

    public Employee(string name, int age, decimal salary, string department, string position)
    {
        Name = name;
        Age = age;
        Salary = salary;
        Department = department;
        Position = position;
    }

    public virtual void GetInfo()
    {
        Console.WriteLine($"Сотрудник {Name}, возраст {Age}, зарплата {Salary} рублей, отдел {Department}, должность {Position}");
    }

    public virtual void Work()
    {
        Console.WriteLine("Сотрудник работает");
    }

    public virtual void TakeVacation()
    {
        Console.WriteLine("Сотрудник берет отпуск");
    }

    public void GetDepartmentInfo()
    {
        Console.WriteLine($"Отдел: {Department}");
    }

    public void Promote(string newPosition)
    {
        Position = newPosition;
        Console.WriteLine($"{Name} повышен на должность {Position}");
    }
}

// Класс Manager, наследующийся от Employee
class Manager : Employee
{
    public int SubordinatesCount { get; set; }
    public string OfficeLocation { get; set; }

    public Manager(string name, int age, decimal salary, string department, string position, int subordinatesCount, string officeLocation) 
        : base(name, age, salary, department, position)
    {
        SubordinatesCount = subordinatesCount;
        OfficeLocation = officeLocation;
    }

    public override void Work()
    {
        Console.WriteLine("Управляет командой специалистов");
    }

    public void ManageTeam(Employee[] employees)
    {
        Console.WriteLine($"Менеджер {Name} управляет командой из {SubordinatesCount} человек:");
        foreach (var employee in employees)
        {
            employee.GetInfo();
        }
    }

    public void OrganizeMeeting()
    {
        Console.WriteLine($"Менеджер {Name} организует встречу в офисе {OfficeLocation}");
    }
}

// Класс Developer, наследующийся от Employee
class Developer : Employee
{
    public string Specialization { get; set; }
    public string CurrentProject { get; set; }

    public Developer(string name, int age, decimal salary, string department, string position, string specialization, string currentProject) 
        : base(name, age, salary, department, position)
    {
        Specialization = specialization;
        CurrentProject = currentProject;
    }

    public override void Work()
    {
        Console.WriteLine("Разработчик работает");
    }

    public override void TakeVacation()
    {
        Console.WriteLine("Есть необходимость согласования отпуска с проектными сроками");
    }

    public void CodeReview()
    {
        Console.WriteLine($"Разработчик {Name} проводит код-ревью для проекта {CurrentProject}");
    }
}


// Класс SeniorDeveloper, наследующийся от Developer (сложное наследование)
class SeniorDeveloper : Developer
{
    public string TeamLead { get; set; }

    public SeniorDeveloper(string name, int age, decimal salary, string department, string position, string specialization, string currentProject, string teamLead) 
        : base(name, age, salary, department, position, specialization, currentProject)
    {
        TeamLead = teamLead;
    }

    public void LeadTeam()
    {
        Console.WriteLine($"Старший разработчик {Name} руководит командой под руководством {TeamLead}");
    }
}

// Интерфейс IReviewable (для множественного наследования)
interface IReviewable
{
    void ReviewWork();
}


// Класс Designer, реализующий интерфейс IReviewable
class Designer : Employee, IReviewable
{
    public string DesignStyle { get; set; }
    public string PreferredTools { get; set; }

    public Designer(string name, int age, decimal salary, string department, string position, string designStyle, string preferredTools) 
        : base(name, age, salary, department, position)
    {
        DesignStyle = designStyle;
        PreferredTools = preferredTools;
    }

    public override void Work()
    {
        Console.WriteLine("Ведется работа над дизайн-проектами");
    }

    public void PresentDesign()
    {
        Console.WriteLine($"Дизайнер {Name} представляет дизайн в стиле {DesignStyle} с использованием инструментов {PreferredTools}");
    }

    public void ReviewWork()
    {
        Console.WriteLine($"Дизайнер {Name} оценивает дизайн-проект в стиле {DesignStyle}");
    }
}

Manager myManager = new Manager("Егор", 27, 500000, "IT", "Менеджер", 2, "Офис 101");
Developer myDeveloper = new Developer("Андрей", 30, 80000, "IT", "Разработчик", "Фронтенд", "Проект X");
Designer myDesigner = new Designer("Мария", 25, 70000, "Design", "Дизайнер", "Минимализм", "Adobe XD");

Employee[] team = { myDeveloper, myDesigner };

myManager.Work();
myManager.TakeVacation();
myManager.GetInfo();
myManager.ManageTeam(team);
myManager.OrganizeMeeting();

myDeveloper.Work();
myDeveloper.TakeVacation();
myDeveloper.GetInfo();
myDeveloper.CodeReview();

myDesigner.Work();
myDesigner.TakeVacation();
myDesigner.GetInfo();
myDesigner.PresentDesign();
((IReviewable)myDesigner).ReviewWork();

// Демонстрация SeniorDeveloper (сложное наследование)
SeniorDeveloper seniorDeveloper = new SeniorDeveloper("Иван", 35, 120000, "IT", "Старший разработчик", "Бэкенд", "Проект Y", "Андрей");
seniorDeveloper.Work();
seniorDeveloper.TakeVacation();
seniorDeveloper.GetInfo();
seniorDeveloper.LeadTeam();
```
