Описание задачи:
Создать базовый класс Employee в C#, который будет представлять информацию о
сотрудниках компании. На основе этого класса разработать 2-3 производных
класса, демонстрирующих принципы наследования и полиморфизма. В каждом из
классов должны быть реализованы новые атрибуты и методы, а также
переопределены некоторые методы базового класса для демонстрации
полиморфизма.
Требования к базовому классу Employee:
• Атрибуты: Имя (Name), Возраст (Age), Зарплата (Salary).
• Методы:
o GetInfo(): метод для получения информации о сотруднике в виде
строки.
o Work(): метод для вывода сообщения о выполнении рабочих
обязанностей.
o TakeVacation(): метод для вывода сообщения о взятии отпуска.

Требования к производным классам:
1. Менеджер (Manager): Должен содержать дополнительные атрибуты, такие
как Количество подчиненных (SubordinatesCount). Метод Work() должен быть
переопределен для добавления информации о управлении командой.
2. Разработчик (Developer): Должен содержать дополнительные атрибуты,
такие как Специализация (Specialization). Метод TakeVacation() должен быть
переопределен для добавления информации о необходимости согласования
отпуска с проектными сроками.
3. Дизайнер (Designer) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Стиль дизайна (DesignStyle).
Метод Work() должен быть переопределен для добавления информации о
работе над дизайн-проектами.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) исользуйтие в проекте коллекции, делегаты, события.

```
using System;

// Базовый класс Employee
public class Employee
{
    private readonly INotificationService _notificationService;
    private string name;
    private int age;
    private decimal salary;
    private string department;
    private string position;
    private DateTime hireDate;
    private string email;
    private bool isActive;
    // Новые атрибуты
    public string PhoneNumber {get; set;}
    public string Address {get; set;}

    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    public int Age
    {
        get { return age; }
        set { age = value; }
    }

    public decimal Salary
    {
        get { return salary; }
        set { salary = value; }
    }

    public string Department
    {
        get { return department; }
        set { department = value; }
    }

    public string Position
    {
        get { return position; }
        set { position = value; }
    }

    public DateTime HireDate
    {
        get { return hireDate; }
        set { hireDate = value; }
    }

    public string Email
    {
        get { return email; }
        set { email = value; }
    }

    public bool IsActive
    {
        get { return isActive; }
        set { isActive = value; }
    }



    public Employee(string name, int age, decimal salary, string department, string position, DateTime hireDate, string email, bool isActive, INotificationService notificationService)
    {
        Name = name;
        Age = age;
        Salary = salary;
        Department = department;
        Position = position;
        HireDate = hireDate;
        Email = email;
        IsActive = true; // По умолчанию сотрудник активен
        _notificationService = notificationService;
    }

    public virtual void GetInfo()
    {
        Console.WriteLine($"Сотрудник {Name}, возраст {Age}, зарплата {Salary} рублей, отдел {Department}, должность {Position}");
    }

    // Перегруженные методы Work
    public virtual void Work()
    {
        Console.WriteLine("Сотрудник работает");
    }

    public virtual void Work(string task) // Перегрузка метода Work
    {
        Console.WriteLine($"Сотрудник работает над задачей: {task}");
    }

    public virtual void Work(int hours) // Перегрузка метода Work
    {
        Console.WriteLine($"Сотрудник работает {hours} часов");
    }

    // Новые методы
    public void SendEmail(string message)
    {
        Console.WriteLine($"Email отправлен на {Email}: {message}");
    }

    public int CalculateYearsOfService()
    {
        return DateTime.Now.Year - HireDate.Year;
    }

    public void Deactivate()
    {
        IsActive = false;
        Console.WriteLine($"{Name} деактивирован");
    }

    public virtual void TakeVacation()
    {
        Console.WriteLine("Сотрудник берет отпуск");
    }

    public void GetDepartmentInfo()
    {
        Console.WriteLine($"Отдел: {Department}");
    }

    public void Promote(string newPosition)
    {
        Position = newPosition;
        Console.WriteLine($"{Name} повышен на должность {Position}");
    }
// Новые методы
    public void SendNotification(string message)
    {
        _notificationService.SendNotification(message);
    }

    public void RequestSickLeave()
    {
        Console.WriteLine($"{Name} запросил больничный.");
    }

    public void UpdateContactInfo(string newPhone, string newEmail)
    {
        PhoneNumber = newPhone;
        Email = newEmail;
        Console.WriteLine($"Контактные данные обновлены для {Name}");
    }
}

// Generic-класс 
public class EmployeeCollection<T> where T : Employee
{
    private List<T> _employees = new List<T>();

    public void AddEmployee(T employee)
    {
        _employees.Add(employee);
        Console.WriteLine($"Добавлен сотрудник: {employee.Name}");
    }

    public void DisplayAll()
    {
        Console.WriteLine("\nСписок всех сотрудников:");
        foreach (var emp in _employees)
        {
            emp.GetInfo();
        }
    }

    public void SendMassEmail(string message)
    {
        Console.WriteLine($"\nРассылка email для {_employees.Count} сотрудников:");
        foreach (var emp in _employees)
        {
            emp.SendEmail(message);
        }
    }
}

// Класс Manager, наследующийся от Employee
class Manager : Employee
{
    public int SubordinatesCount { get; set; }
    public string OfficeLocation { get; set; }

    public Manager(string name, int age, decimal salary, string department, string position, int subordinatesCount, string officeLocation, DateTime hireDate, string email, bool isActive, INotificationService _notificationService) 
        : base(name, age, salary, department, position, hireDate, email, isActive, _notificationService)
    {
        SubordinatesCount = subordinatesCount;
        OfficeLocation = officeLocation;
    }

    public override void Work()
    {
        Console.WriteLine("Управляет командой специалистов");
    }

    public void ManageTeam(Employee[] employees)
    {
        Console.WriteLine($"Менеджер {Name} управляет командой из {SubordinatesCount} человек:");
        foreach (var employee in employees)
        {
            employee.GetInfo();
        }
    }

    public void OrganizeMeeting()
    {
        Console.WriteLine($"Менеджер {Name} организует встречу в офисе {OfficeLocation}");
    }

    // Новый метод 
    public void ScheduleMeeting(string topic)
    {
        Console.WriteLine($"Запланирована встреча по теме: {topic} в {OfficeLocation}");
    }
}

// Класс Developer, наследующийся от Employee
class Developer : Employee
{
    public string Specialization { get; set; }
    public string CurrentProject { get; set; }
    public bool IsRemote { get; set; } //Новый атрибут

    public Developer(string name, int age, decimal salary, string department, string position, string specialization, string currentProject, bool isRemote, DateTime hireDate, string email, bool isActive, INotificationService _notificationService) 
        : base(name, age, salary, department, position, hireDate, email, isActive, _notificationService)
    {
        Specialization = specialization;
        CurrentProject = currentProject;
        IsRemote = isRemote;
    }

    public override void Work()
    {
        Console.WriteLine("Разработчик работает");
    }

    public override void TakeVacation()
    {
        Console.WriteLine("Есть необходимость согласования отпуска с проектными сроками");
    }

    public void CodeReview()
    {
        Console.WriteLine($"Разработчик {Name} проводит код-ревью для проекта {CurrentProject}");
    }

    public void FixBug(string bugId) //Новый метод
    {
        Console.WriteLine($"Исправление бага {bugId} в проекте {CurrentProject}");
    }

    public void DeployCode(string environment) //Новый метод
    {
        Console.WriteLine($"{Name} деплоит код в {environment}");
    }
}


// Класс SeniorDeveloper, наследующийся от Developer (сложное наследование)
class SeniorDeveloper : Developer
{
    public string TeamLead { get; set; }

    public SeniorDeveloper(string name, int age, decimal salary, string department, string position, string specialization, string currentProject, string teamLead, DateTime hireDate, string email, bool isActive, bool isRemote, INotificationService _notificationService) 
        : base(name, age, salary, department, position, specialization, currentProject, isRemote, hireDate, email, isActive, _notificationService)
    {
        TeamLead = teamLead;
    }

    public void LeadTeam()
    {
        Console.WriteLine($"Старший разработчик {Name} руководит командой под руководством {TeamLead}");
    }
}

// Интерфейс IReviewable (для множественного наследования)
interface IReviewable
{
    void ReviewWork();
}


// Класс Designer, реализующий интерфейс IReviewable
class Designer : Employee, IReviewable
{
    public string DesignStyle { get; set; }
    public string PreferredTools { get; set; }
    public int ProjectsCompleted { get; set; } //Новый

    public Designer(string name, int age, decimal salary, string department, string position, string designStyle, string preferredTools, DateTime hireDate, string email, bool isActive, INotificationService _notificationService) 
        : base(name, age, salary, department, position, hireDate, email, isActive, _notificationService)
    {
        DesignStyle = designStyle;
        PreferredTools = preferredTools;
    }

    public override void Work()
    {
        Console.WriteLine("Ведется работа над дизайн-проектами");
    }

    public void PresentDesign()
    {
        Console.WriteLine($"Дизайнер {Name} представляет дизайн в стиле {DesignStyle} с использованием инструментов {PreferredTools}");
    }

    void IReviewable.ReviewWork()
    {
        Console.WriteLine($"(Явная реализация) Дизайнер {Name} проверяет проект.");
    }

    public void ReviewWork()
    {
        Console.WriteLine($"Дизайнер {Name} оценивает дизайн-проект в стиле {DesignStyle}");
    }

    public void CelebrateMilestone() //Новый
    {
        Console.WriteLine($"{Name} завершил {ProjectsCompleted} проектов!");
    }

    
}

// Интерфейс сервиса уведомлений
public interface INotificationService
{
    void SendNotification(string message);
}

// Реализация сервиса
public class EmailNotificationService : INotificationService
{
    public void SendNotification(string message)
    {
        Console.WriteLine($"Отправлено уведомление: {message}");
    }
}

// Создаём сервис уведомлений
INotificationService notificationService = new EmailNotificationService();


Manager myManager = new Manager("Егор", 27, 500000, "IT", "Менеджер", 2, "Офис 101", new DateTime(2020, 5, 10), "egor@company.com", true, notificationService);
myManager.SendNotification("Планёрка в 18:00");
Developer myDeveloper = new Developer("Андрей", 30, 80000, "IT", "Разработчик", "Фронтенд", "Проект X", true, new DateTime(2020, 6, 10), "andrey@company.com", true, notificationService);
Designer myDesigner = new Designer("Мария", 25, 70000, "Design", "Дизайнер", "Минимализм", "Adobe XD", new DateTime(2020, 6, 10), "maria@company.com", true, notificationService);

// Создаем коллекцию сотрудников
EmployeeCollection<Employee> company = new EmployeeCollection<Employee>();
company.AddEmployee(myManager);
company.AddEmployee(myDeveloper);
company.AddEmployee(myDesigner);

Employee[] team = { myDeveloper, myDesigner };

myManager.Work();
myManager.Work("Планирование спринта");
myManager.Work(8);
myManager.TakeVacation();
myManager.GetInfo();
myManager.ManageTeam(team);
myManager.OrganizeMeeting();

myDeveloper.Work();
myDeveloper.TakeVacation();
myDeveloper.GetInfo();
myDeveloper.CodeReview();

myDesigner.Work();
myDesigner.TakeVacation();
myDesigner.GetInfo();
myDesigner.PresentDesign();
((IReviewable)myDesigner).ReviewWork();

// Демонстрация SeniorDeveloper (сложное наследование)
SeniorDeveloper seniorDeveloper = new SeniorDeveloper("Иван", 35, 120000, "IT", "Старший разработчик", "Бэкенд", "Проект Y", "Андрей", new DateTime(2020, 4, 10), "ivan@company.com", true, true, notificationService);
seniorDeveloper.Work();
seniorDeveloper.TakeVacation();
seniorDeveloper.GetInfo();
seniorDeveloper.LeadTeam();
```
